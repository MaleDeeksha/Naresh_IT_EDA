**method-1**
  - first consider one categorical column
  - observe how many unique labels are there
  - assume here we are taking case_status column
  - case_status two unique labels are there
    - denied
    - certified
  - create a dictionary by providing key:value
  - here the key= denied and the value=0
  - key=certified and the value=0
  - dictionary will looks:{'certified':0,'denied':1}  

```python
visa_df['case_status'].unique()
```

```python
d1={'Certified':0,'Denied':1}
visa_df['case_status'].map(d1)
visa_df['case_status']=visa_df['case_status'].map(d1)
```

```python
visa_df.head()
```

```python
d1={}
lables=visa_df['continent'].unique()
for i in range(len(lables)):
    d1[lables[i]]=i
visa_df['case_status']=visa_df['case_status'].map(d1)
visa_df.head() 
```

**np-where**
- np where is a kind of if-else
- np where is useful for binary condition
- which means it is useful for the column having two labels
- np whrer will take 3 parameters
  - condition
  - true value
  - false value
- i want replace case status column with 1 when it is denied
- otherwise it is zero
- con=visa_df['case_status']=='denied'
- true value=1 false value=0
- np.where(con,true value,false value)

```python
con=visa_df['case_status']=='Denied'
visa_df['case_status']=np.where(con,1,0)
visa_df.head()
```

```python
**method-3$**
$one-hot-encoder$
pd.getdummies()
- one hot encoder means at a time only one will be ON ,other is OFF
- ON represents with 1 and OFF represents with 0
- one hot encoder creates new extra columns based on labels in the column
- for example case_status has two unique labels
- if you apply one hot encoder on case_status i creates two extra column
  - case_status_denied
  - case_status_certified


```

```python
**advantage**
- one hot encoder means at atime one will be ON other is OFF
- it is one of the important property before apply ml models
- variable are independent each other
- variable are perpenducular each other
- variable are orthogonal each other
- variable are 90 degrees phase each other
**disadvantages**
- assume that column has 100 unique labels
- if we apply one hot encoder it will create 100 new column
- your data matrix become sparse
- we required more time and more memory to compute the calculations
- this is called curse of dimensionality
```
